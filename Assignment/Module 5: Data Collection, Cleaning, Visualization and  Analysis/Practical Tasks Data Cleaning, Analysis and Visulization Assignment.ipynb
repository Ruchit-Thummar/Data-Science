# Questions

```python
# Q-1 Which cars brand sells the cheapest convirtible model.

data[data['Vehicle Style'].str.contains('Convertible')].sort_values('MSRP').head(1)[['Make','Vehicle Size','MSRP']]
```

```python
# Q-2 Which year introduced the most new electric car?
data[data['Engine Fuel Type'].str.contains("electric",na=False)].groupby('Year').size().sort_values(ascending=False).head(1)
```

```python
# Q-3 What is the most expensive 4dr SUV in the dataset?
data[data['Vehicle Style'].str.contains('4dr SUV')].sort_values('MSRP',ascending=False).head(1)
```

```python
# Q-4 For each year, show me the first car in the dataset that had more than 600 HP.

data[data['Engine HP'] > 600].groupby('Year').first()
```

```python
# Q-5 Which car has the best MPG per horsepower.
data.iloc[(data['highway MPG'] / data['Engine HP']).argmax()][['Make']]
```

```python
#  Q-6 Which vehicle size shows the highest variety of engine cylinder?
data.groupby('Vehicle Size')['Engine Cylinders'].nunique()
```

```python
#  Q-7 which is the most comman vehicale size among Convertible cars.

data[data['Vehicle Style'] == 'Convertible'].groupby('Vehicle Size').size().sort_values(ascending=False).head(1)
```

```python
# Q-8 which brand sold the car with the lowest horsepower.

data.loc[data['Engine HP'].idxmin(), ['Make','Model','Engine HP']]
```

```python
# Q-9 which car with more than 8 cylinders had the chepest price?

data[data['Engine Cylinders'] > 8.0].groupby('MSRP').min().head(1)
```

```python
# Q-10 Which year had the overall lowest average horsepower?

data.groupby('Year')['Engine HP'].mean().idxmin()

```

## DATA VISULIZATION

```python
# What is the fastest-growing fuel type over the years

fuel = data.groupby(['Year', 'Engine Fuel Type']).size().unstack(fill_value=0)
plt.figure(figsize=(12,6))
fuel.plot(kind='line', marker='o',figsize=(12,6))
plt.title("Fuel Type Popularity Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Cars")
plt.grid()


```

```python
#which size of vehicles get the highest highway MPG

mpg = data.groupby('Vehicle Size')['highway MPG'].max()
plt.figure(figsize=(9,6))
mpg.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Maximum Highway MPG by Vehicle Size")
plt.xlabel("Vehicle Size")
plt.ylabel("Highway MPG")
plt.xticks(rotation=0)
plt.show()
```

```python
# Find the most comman Transmission Types
plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='Transmission Type', palette="pastel",hue='Transmission Type',edgecolor='black')
plt.title("Most Common Transmission Types")
plt.xlabel("Transmission Type")
plt.ylabel("Count")
plt.xticks(rotation=10)
plt.show()
```

```python
# What is the average MSRP for each Vehicle Style?
d1 = data.groupby('Vehicle Style')['MSRP'].mean().reset_index().sort_values('MSRP', ascending=False)
plt.figure(figsize=(12, 7))
sns.barplot(data=d1,x='Vehicle Style',y='MSRP',palette="viridis")
plt.title("Average MSRP by Vehicle Style")
plt.xlabel("Vehicle Style")
plt.ylabel("Average MSRP")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# find the distribution of Engine HP.
plt.figure(figsize=(10, 6))
sns.histplot(data['Engine HP'], kde=True, color='purple')
plt.title("Distribution of Engine Horsepower")
plt.xlabel("Engine HP")
plt.ylabel("Frequency")
```
