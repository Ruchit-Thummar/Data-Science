a. Find the mean, median, and mode

4. Use the “Student Attitude and Behavior” dataset and select the following columns: Height,
Weight, 10th Marks, 12th Marks.

• Find Mean, Median, Mode and Quartiles

• Find Skewness, kurtosis and Interpretation Based on the Skewness, kurtosis.


• Find Mean, Median, Mode and Quartiles


```python
cols = st[['Height(CM)','Weight(KG)','10th Mark','12th Mark']]
mean1 = cols.mean()
median1 = cols.median()
mode1 = stats.mode(cols)

print('Mean',mean1,'\n\nMedian',median1,'\n\nMode',mode1,'\n')

columns = ['Height(CM)','Weight(KG)','10th Mark','12th Mark']

q1 = np.quantile(cols,0.25,axis=0)
q2 = np.quantile(cols,0.5,axis=0)
q3 = np.quantile(cols,0.75,axis=0)

for i, name in enumerate(columns):
  print(f'{name}--> Q1={q1[i]}, Q2={q2[i]}, Q3={q3[i]}')
```


• Find Skewness, kurtosis and Interpretation Based on the Skewness, kurtosis.

```python
skewn = cols.skew()
kurto = cols.kurtosis()

print('Skewness',skewn,'\n\nKurtosis',kurto)
```

```python
skewn.plot(kind='bar', color='skyblue')
plt.title('Skewness of columns')
plt.ylabel('Skewness')
plt.xticks(rotation=0)
plt.show()
```

```python
kurto.plot(kind='bar', color='lightgreen')
plt.title('Kurtosis of columns')
plt.ylabel('Kurtosis')
plt.xticks(rotation=0)
plt.show()
```

Interpretation Based on the Skewness, kurtosis.

--> Height and 10th Mark are left-skewed and higher from the centre, which means they are highly asymmetric.

--> Weight and 12th Mark are mostly symmetric and closer to a normal distribution.

5. An investment analyst wants to assess the relationship between the stock prices of two
companies to identify potential investment opportunities. Data: Let's consider the daily
closing prices (in dollars) of Company A and Company B for a sample of 20 trading days:

Company A: 45, 47, 48, 50, 52, 53, 55, 56, 58, 60, 62, 64, 65, 67, 69, 70, 72, 74, 76, 77

Company B: 52, 54, 55, 57, 59, 60, 61, 62, 64, 66, 67, 69, 71, 73, 74, 76, 78, 80, 82, 83

• Calculate the covariance and correlation between the stock prices of Company A
and Company B.

 Interpret the value of the covariance and explain the nature of the
relationship between the two stocks.


• Calculate the covariance and correlation between the stock prices of Company A
and Company B.


```python
comp_a = np.array([45, 47, 48, 50, 52, 53, 55, 56, 58, 60, 62, 64, 65, 67, 69, 70, 72, 74, 76, 77])
comp_b = np.array([52, 54, 55, 57, 59, 60, 61, 62, 64, 66, 67, 69, 71, 73, 74, 76, 78, 80, 82, 83])
```

```python
covariance = np.cov(comp_a,comp_b)
print(covariance)
```

```python
correlation = np.corrcoef(comp_a,comp_b)
print(correlation[0,1])
```

```python
plt.plot(comp_a, label='Company A')
plt.plot(comp_b, label='Company B')
plt.xlabel('Day')
plt.ylabel('Price')
plt.title('Stock Prices')
plt.legend()
plt.show()
```

Interpret the value of the covariance and explain the nature of the relationship between the two stocks.

--> The graph shows Company A and B stock prices over 20 days, both stocks are going up overtime.

--> The high correlation is 0.998 that confirms two stocks are strongly related.
